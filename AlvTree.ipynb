```python

#ARVORE


#              4
#             /  \
#            /    \
#           2      6
#          / \
#         /   \
#        1     3

class No:

  valor = None
  f_esq = None
  f_dir = None
  
 
  def __init__(self, valor = None):
    self.valor = valor
    self.setaFilhos(None, None)

  def setaFilhos(self, esquerda, direita):
   self.f_esq = esquerda
   self.f_dir = direita
    

  def __str__(self):
    return "No:" + str(self.valor)


class AvlTree:
  
  def __init__(self, data = None):
    no = No(data)
    self.raiz = no
    


  #Percurso inorder (Percurso em pre-ordem; esquerda, raiz e direita)
  def simetric_traversal(self, no = None):
    if no is None:
      no = self.raiz
    if no.f_esq:
      self.simetric_traversal(no.f_esq)
    print(no ," ", end="")
    if no.f_dir:
     self.simetric_traversal(no.f_dir)
 
  #Percurso inorder (Percurso em pos-ordem; esquerda, direita, e raiz)
  def postorder_traversal(self, no = None):
    if no is None:
     no = self.raiz
    if no.f_esq:
      self.postorder_traversal(no.f_esq)
    if no.f_dir:
      self.postorder_traversal(no.f_dir)
      print(no ,"", end="")

  #Imprime apenas os valores das folhas
  def valor_folhas(self, no = None): 
    #verifica se o No é None
    if no is None:
      no = self.raiz
    #Verifica se o No tem filho a esquerda
    if no.f_esq:
      #chama a função novamente passando a nova raiz
      self.valor_folhas(no.f_esq)
    
    #Verifica se o No tem filho a direita
    if no.f_dir:
     #chama a função novamente passando a nova raiz
     self.valor_folhas(no.f_dir)
    #Verifica se o No tem os filhos definido como None, se sim exibe o No(folha)
    if no.f_esq == None and no.f_dir == None:
      print(no)

  def balanco(self, no):
     
    prof_esq = 0
    
    if no.f_esq:
      prof_esq = self.profundidade(no.f_esq)
    
    prof_dir = 0
    if no.f_dir:
      prof_dir = self.profundidade(no.f_dir)

    return prof_esq - prof_dir
  
  def profundidade(self, no):
    prof_esq = 0
    
    if no.f_esq:
      prof_esq = self.profundidade(no.f_esq)
    
    prof_dir = 0
    if no.f_dir:
      prof_dir = self.profundidade(no.f_dir)
    
    return 1 + max(prof_esq, prof_dir) 
  
  #ROTAÇÃO SIMPLES PARA DIREITA
  def RSD(self, no): 

    no.valor, no.f_esq.valor = no.f_esq.valor, no.valor
    old_dir = no.f_dir 

    no.setaFilhos(no.f_esq.f_esq, no.f_esq)
    no.f_dir.setaFilhos(no.f_dir.f_dir, old_dir)
  
  #ROTAÇÃO SIMPLES PARA ESQUERDA
  def RSE(self, no):
    no.valor, no.f_dir.valor = no.f_dir.valor, no.valor
    old_esq = no.f_esq 

    no.setaFilhos(no.f_dir, no.f_dir.f_dir)
    no.f_esq.setaFilhos(old_esq, no.f_esq.f_esq)

  def rotacaoDireitaEsquerda(self, no):
    self.RSD(no.f_dir)
    self.RSE(no)

  def rotacaoEsquerdaDireita(self, no):
   self.RSE(no.f_esq)
   self.RSD(no)

  def executaBalanco(self, no):
    bal = self.balanco(no)

    if bal > 1:
      if self.balanco(no.f_esq) > 0:
        self.RSD(no)
      else:
        self.rotacaoEsquerdaDireita(no)
    elif bal < -1:
      if self.balanco(no.f_esq) < 0:
        self.RSE(no)
      else: 
        self.rotacaoDireitaEsquerda(no)#chamou1

  def insere(self, data):
    return self.__insere(data, self.raiz)

  def __insere(self, data, no):
    if no.valor == None:
      self.raiz = No(data)
      return

    if data <= no.valor:
      if not no.f_esq:
       no.f_esq = No(data)
      else:
         self.__insere(data, no.f_esq)
    else:
       if not no.f_dir:
        no.f_dir = No(data)
       else:
         self.__insere(data, no.f_dir)
    self.executaBalanco(no)


  def imprimeArvore(self, no = None):
      if no is None:
        no = self.raiz
      print(no)
      if no.f_esq:
        self.imprimeArvore(no.f_esq)
      if no.f_dir:
        self.imprimeArvore(no.f_dir)
  
    

  
  
  def altura(self, no = None):
    if no is None:
     no = self.raiz

    hLeft = 0
    hRight = 0
    
    if no.f_esq:
      hLeft = self.altura(no.f_esq)
    
    if no.f_dir:
     hRight = self.altura(no.f_dir)
    
    if (hRight > hLeft):
      return hRight +1
   
    return hLeft +1

  def quantidade_nos(self, no = None):
    if no is None:
      no = self.raiz

    count = 0
    if no.f_esq:
      count += self.quantidade_nos(no.f_esq)
    if no.f_dir:
      count += self.quantidade_nos(no.f_dir)
    
    return count +1
    
  #Busca um valor X na arvore
  def buscar(self, x, no = None):
    
    #verifica se o No é None
    if no is None:
      no = self.raiz
   
    
    #Verifica se são iguais, se sim imprime e retorna o No
    if no.valor == x:
      return no
    
    ret = None
      #Verifica se o No tem filho a esquerda
    if no.f_esq:
        #chama a função novamente passando o x e a nova raiz
       ret = self.buscar(x, no.f_esq)
    
      #Verifica se o No tem filho a direita
    if no.f_dir and ret == None:
      #chama a função novamente passando o x e a nova raiz
       ret = self.buscar(x, no.f_dir)
   
    return ret
  
  def inserir(self, valor, no = None):
   
    if no is None:
      no = self.raiz
      
     
    #verifica se o valor é menor do que o no, se for menor segue para o lado esquerdo
    if valor < no.valor:
      #Verifica se o No à esquerda está definido
      if no.f_esq:
        #chama a função novamente passando a nova raiz
        self.inserir(valor, no.f_esq)
      #caso contrario, o filho a esquerda recebe o valor
      else:
        no.f_esq = No(valor)
    
    #caso contrario, segue para o lado direito    
    else:
      #Verifica se o No à direita está definido
      if no.f_dir:
        #chama a função novamente passando a nova raiz
        self.inserir(valor, no.f_dir)
      #caso contrario, o filho a direita recebe o valor
      else:
        no.f_dir = No(valor)
    
  def insert(self, valor):
    return self._insert(valor, self.raiz)
    
  def _insert(self, valor, no):
    if(no.valor == None):
      no.valor = valor
      no.f_esq = No()
      no.f_dir = No()
      return

    if(valor < no.valor):
      self._insert(valor, no.f_esq)
    
    elif(valor > no.valor):
      self._insert(valor, no.f_dir)
  
  def red(self, no = None):
    if no is None:
      no = self.raiz
    
    print(no,"(", end="")
    
    if no.f_esq:  
      self.red(no.f_esq)
     
    if no.f_dir:
      self.red(no.f_dir)
      
    print(")", end="")

  def min(self, no = None):
        if no == None:
            no = self.raiz
        while no.f_esq:
            no = no.f_esq
        return no.valor
  
  def remove(self, value, no = None):
    if no is None:
      no = self.raiz
    
    if value < no.valor:
      no.f_esq = self.remove(value, no.f_esq)
    elif value > no.valor:
      no.f_dir = self.remove(value, no.f_dir)  
    else:
      if no.f_esq is None:
        return no.f_dir
      
      elif no.f_dir is None:
        return no.f_esq
      
      else:
        substituto = self.min(no.f_dir)
        no.valor = substituto
        no.f_dir = self.remove(substituto, no.f_dir)
    
    return no
  
    


tree = AvlTree()
tree.insere(6)
tree.insere(2)
tree.insere(1)
tree.insere(9)
tree.insere(3)
tree.insere(8)


tree.imprimeArvore()


#print(tree.raiz) #6
#print(tree.raiz.f_esq) #2
#print(tree.raiz.f_esq.f_esq) #1
#print(tree.raiz.f_esq.f_dir) #3
#print(tree.raiz.f_dir) #9
#print(tree.raiz.f_dir.f_esq) #8
#print(tree.raiz.f_dir.f_esq.f_esq) #7


#print("fator de balanco: ")
#print(tree.balanco(tree.raiz.f_dir))

#tree.RSD(tree.raiz.f_dir)
#print("dps da rotação: ")
#print(tree.raiz) #6
#print(tree.raiz.f_esq) #2
#print(tree.raiz.f_esq.f_esq) #1
#print(tree.raiz.f_esq.f_dir) #3
#print(tree.raiz.f_dir) #8
#print(tree.raiz.f_dir.f_esq) #7
#print(tree.raiz.f_dir.f_dir) #9

#print("fator de balanco: ")
#print(tree.balanco(tree.raiz.f_dir))

#tree.simetric_traversal()

#tree.valor_folhas()
#print(tree.buscar(3))
#print("altura: ", tree.altura())
#print("Quantidade de nos: " ,tree.quantidade_nos())
#tree.postorder_traversal()
#tree.inserir(5)

#tree.red()
#tree.simetric_traversal()
#print("antes de remover: ",end="") 
#tree.simetric_traversal()
#tree.remove(3)
#print("\nApós remover: ", end="")
#tree.simetric_traversal()

#print("\nMinimo: ", tree.min())
```

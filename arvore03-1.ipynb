```python
#ARVORE


#              4
#             /  \
#            /    \
#           2      6
#          / \
#         /   \
#        1     3

class No:
 
  def __init__(self,valor):
    self.valor = valor
    self.f_esq = None
    self.f_dir = None
    

  def __str__(self):
    return str(self.valor)


class BinaryTree:
  
  def __init__(self, data):
    no = No(data)
    self.raiz = no
    

  #Percurso inorder (Percurso em pre-ordem; esquerda, raiz e direita)
  def simetric_traversal(self, no = None):
    if no is None:
      no = self.raiz
    if no.f_esq:
      self.simetric_traversal(no.f_esq)
    print(no ," ", end="")
    if no.f_dir:
     self.simetric_traversal(no.f_dir)
 
  #Percurso inorder (Percurso em pos-ordem; esquerda, direita, e raiz)
  def postorder_traversal(self, no = None):
    if no is None:
     no = self.raiz
    if no.f_esq:
      self.postorder_traversal(no.f_esq)
    if no.f_dir:
     self.postorder_traversal(no.f_dir)
     print(no ,"", end="")

  #Imprime apenas os valores das folhas
  def valor_folhas(self, no = None): 
    #verifica se o No é None
    if no is None:
      no = self.raiz
    #Verifica se o No tem filho a esquerda
    if no.f_esq:
      #chama a função novamente passando a nova raiz
      self.valor_folhas(no.f_esq)
    
    #Verifica se o No tem filho a direita
    if no.f_dir:
     #chama a função novamente passando a nova raiz
     self.valor_folhas(no.f_dir)
    #Verifica se o No tem os filhos definido como None, se sim exibe o No(folha)
    if no.f_esq == None and no.f_dir == None:
      print(no)
  
  def altura(self, no = None):
    if no is None:
     no = self.raiz

    hLeft = 0
    hRight = 0
    
    if no.f_esq:
      hLeft = self.altura(no.f_esq)
    
    if no.f_dir:
     hRight = self.altura(no.f_dir)
    
    if (hRight > hLeft):
      return hRight +1
   
    return hLeft +1

  def quantidade_nos(self, no = None):
    if no is None:
      no = self.raiz

    count = 0
    if no.f_esq:
      count += self.quantidade_nos(no.f_esq)
    if no.f_dir:
      count += self.quantidade_nos(no.f_dir)
    
    return count +1
    
  #Busca um valor X na arvore
  def buscar(self, x, no = None):
    
    #verifica se o No é None
    if no is None:
      no = self.raiz
   
    
    #Verifica se são iguais, se sim imprime e retorna o No
    if no.valor == x:
      return no
    
    ret = None
      #Verifica se o No tem filho a esquerda
    if no.f_esq:
        #chama a função novamente passando o x e a nova raiz
       ret = self.buscar(x, no.f_esq)
    
      #Verifica se o No tem filho a direita
    if no.f_dir and ret == None:
      #chama a função novamente passando o x e a nova raiz
       ret = self.buscar(x, no.f_dir)
   
    return ret
  
  def inserir(self, valor, no = None):
    if no is None:
      no = self.raiz

    #verifica se o valor é menor do que o no, se for menor segue para o lado esquerdo
    if valor < no.valor:
      #Verifica se o No à esquerda está definido
      if no.f_esq:
        #chama a função novamente passando a nova raiz
        self.inserir(valor, no.f_esq)
      #caso contrario, o filho a esquerda recebe o valor
      else:
        no.f_esq = No(valor)
    
    #caso contrario, segue para o lado direito    
    else:
      #Verifica se o No à direita está definido
      if no.f_dir:
        #chama a função novamente passando a nova raiz
        self.inserir(valor, no.f_dir)
      #caso contrario, o filho a direita recebe o valor
      else:
        no.f_dir = No(valor)
  
  def red(self, no = None):
    if no is None:
      no = self.raiz
    
    print(no,"(", end="")
    
    if no.f_esq:  
      self.red(no.f_esq)
     
    if no.f_dir:
      self.red(no.f_dir)
      
    print(")", end="")

  def min(self, no = None):
        if no == None:
            no = self.raiz
        while no.f_esq:
            no = no.f_esq
        return no.valor
  
  def remove(self, value, no = None):
    if no is None:
      no = self.raiz
    
    if value < no.valor:
      no.f_esq = self.remove(value, no.f_esq)
    elif value > no.valor:
      no.f_dir = self.remove(value, no.f_dir)  
    else:
      if no.f_esq is None:
        return no.f_dir
      
      elif no.f_dir is None:
        return no.f_esq
      
      else:
        substituto = self.min(no.f_dir)
        no.valor = substituto
        no.f_dir = self.remove(substituto, no.f_dir)
    
    return no
  
    


tree = BinaryTree(4)
tree.raiz.f_dir = No(6)
tree.raiz.f_esq = No(2)
tree.raiz.f_esq.f_esq = No(1)
tree.raiz.f_esq.f_dir = No(3)



#tree.simetric_traversal()

#tree.valor_folhas()
#print(tree.buscar(3))
#print("altura: ", tree.altura())
#print("Quantidade de nos: " ,tree.quantidade_nos())
#tree.postorder_traversal()
#tree.inserir(5)
#tree.red()

#print("antes de remover: ",end="") 
#tree.simetric_traversal()
#tree.remove(4)
#print("\nApós remover: ", end="")
#tree.simetric_traversal()

#print("\nMinimo: ", tree.min())
```

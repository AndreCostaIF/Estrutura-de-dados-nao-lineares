```python
#ARVORE


#               RAIZ
#             /      \
#            /        \
#           No         No
#                     /  \
#                    /    \
#                   NO     No

class No:
 
  def __init__(self,valor):
    self.valor = valor
    self.f_esq = None
    self.f_dir = None
    

  def __str__(self):
    return str(self.valor)


class BinaryTree:
  
  def __init__(self, data):
    no = No(data)
    self.raiz = no
    

  #Percurso inorder (Percurso em pre-ordem; esquerda, raiz e direita)
  def simetric_traversal(self, no = None):
    if no is None:
      no = self.raiz
    if no.f_esq:
      self.simetric_traversal(no.f_esq)
    print(no)
    if no.f_dir:
     self.simetric_traversal(no.f_dir)
  #Percurso inorder (Percurso em pos-ordem; esquerda, direita, e raiz)
  def postorder_traversal(self, no = None):
    if no is None:
     no = self.raiz
    if no.f_esq:
      self.simetric_traversal(no.f_esq)
    if no.f_dir:
     self.simetric_traversal(no.f_dir)
     print(no)

  #Imprime apenas os valores das folhas
  def valor_folhas(self, no = None): 
    #verifica se o No é None
    if no is None:
      no = self.raiz
    #Verifica se o No tem filho a esquerda
    if no.f_esq:
      #chama a função novamente passando a nova raiz
      self.valor_folhas(no.f_esq)
    #Verifica se o No tem os filhos definido como None, se sim exibe o No(folha)
    if no.f_esq == None and no.f_dir == None:
      print(no)
    #Verifica se o No tem filho a direita
    if no.f_dir:
     #chama a função novamente passando a nova raiz
     self.valor_folhas(no.f_dir)

  def altura(self, no = None):
    if no is None:
     no = self.raiz
    hLeft = 0
    hRight = 0
    if no.f_esq:
      hLeft = self.altura(no.f_esq)
    if no.f_dir:
     hRight = self.altura(no.f_dir)
    if (hRight > hLeft):
      return hRight +1
    return hLeft +1

  def quantidade_nos(self, no = None):
    if no is None:
      no = self.raiz

    count = 0
    if no.f_esq:
      count += self.quantidade_nos(no.f_esq)
    if no.f_dir:
      count += self.quantidade_nos(no.f_dir)
    
    return count +1
    
  #Busca um valor X na arvore
  def buscar(self, x, no = None):
    
    #verifica se o No é None
    if no is None:
      no = self.raiz
   
    #transforma o No em String
    n = str(no)
    #Transforma String em INT
    n = int(n)
    #Verifica se são iguais, se sim imprime e retorna o No
    if n == x:
      print(no)
      return None
    
    else:
      #Verifica se o No tem filho a esquerda
      if no.f_esq:
        #chama a função novamente passando o x e a nova raiz
        self.buscar(x, no.f_esq)
    
      #Verifica se o No tem filho a direita
      if no.f_dir:
      #chama a função novamente passando o x e a nova raiz
        self.buscar(x, no.f_dir)
      
     

tree = BinaryTree(7)
tree.raiz.f_esq = No(18)
tree.raiz.f_dir = No(14)
tree.raiz.f_dir.f_esq = No(5)
tree.raiz.f_dir.f_dir = No(2)


tree.simetric_traversal()
#tree.valor_folhas()
#tree.buscar(5)
#print("altura: ", tree.altura())
#print("Quantidade de nos: " ,tree.quantidade_nos())

#tree.postorder_traversal()
```
